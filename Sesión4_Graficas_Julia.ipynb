## creación de graficas en Julia

Como hemos visto en las diferentes sesiones, existen diferentes formas para realizar esta tarea: (i) podemos invocar `PyPlot`, (ii) usar `Plots.jl`. En esta sesión utilizaremos algunos comandos revisados en las sesiones pasadas.

### Ejemplo 1

La ley de los gases establece que un proceso de expansión isotermico sigue la relación $\small P=nRT/V$. Graficaremos la curva isotermica de expansión de un 1 mol de un gas ideal desde 4 L hasta 40 L a una temperatura de $\small 298 K$.

```julia
using Plots

V = 4:1:40            # generamos un vector con los datos de V desde 4L hasta 40L con un paso= 0.2
p(V) = (8.314*298)/V  # P (atm)
P= p.(V)              # aplicamos la función p(V) a cada valor de V generado
```

Antes de definir los parámetros de nuestra gráfica, debemos elegir el "Backend" que defien la formade presetnación del grafico dependiendo de que tipod e información manejemos o que tipo de presnetación weuremos usar. Puedes consultar este link para ver diferentes ejemplos: [Backends](https://docs.juliaplots.org/latest/backends/)

```julia
gr()  #Banckend
```

```julia
plot(V,P, label="linea")
scatter!(V,P, label="puntos")
xlabel!("Volumen (L)")
ylabel!("Presión (atm)")
title!("Isoterma a 298 K")
```

Otra forma de obtener la misa misma curva podría ser la siguiente:

```julia
fig = Plots.plot()     
Plots.scatter!(V,P, label="puntos")
Plots.plot!(V,P, label="linea")
Plots.title!("Isoterma a 298 K")
Plots.ylabel!("Presión (atm)")
Plots.xlabel!("Volumen (L)")
display(fig)
```

Para generar varias graficas (subgraficas) dentro de un mismo grafico, debemos definir cada subgrafica individualmente utilizando la función `plot`. Después de defini todas las "subplots" se genera un solografico con todos los elementos definidos

```julia
G1=plot(V,V)   #definimos cad subplot
G2=plot(V,V.^-1)
G3=plot(V,V.^2)
G4=plot(V,V.^3)
G5=plot(V,V.^0.5)
G6=plot(V,V.^4)
plot(G1,G2,G3,G4,G5,G5, layout=(2,3), legend=false)  #definimos la grafica completa
```

## Graficas 3D

Para este ejemplo utilizaremos las siguientes ecuaciones paramétricas:

$\small x=(2+cos(t)) cos( \varphi )$

$\small y=(2 + cos(t))sen( \varphi )$

$\small z=t$

Con $\small t$  $\small \epsilon$  $\small [0, 2 \pi]$, $\small \varphi$  $\small \epsilon$  $\small [0, 2 \pi]$

Primero definimos los vectores con los valores de $\small t$ y $\small \varphi$. Luego las funciones para cada eje

```julia
ts= 0:0.1:2*pi
ps= 0:0.1:2*pi

xs=[2+cos(t)*cos(p) for t=ts, p=ps]
ys=[2+cos(t)sin(p) for t=ts, p=ps]
zs=[t for t=ts]
```

```julia
plot(xs,ys,zs, zlabel="eje z", label=false)
title!("Ecuación paramétrica No.1")
ylabel!("Eje Y")
xlabel!("Eje x")
xlims!(1,3)
ylims!(1,3)
zlims!(0,3)
```

Para este ejemplo utilizaremos las siguientes ecuaciones paramétricas:

$\small x=rcos(t)$

$\small y=rsen(t)$

$\small z=ht$

Con $\small t$  $\small \epsilon$  $\small [0, 6 \pi]$, $\small h=0.4$  $\small r=2.5$

```julia
ts= 0:0.2:6*pi
h=0.4
r=2.5
x1s=[r*cos(t) for t=ts]
y1s=[r*sin(t) for t=ts]
z1s=[h*t for t=ts]
```

```julia
using Plots
scatter!(x1s,y1s,z1s, color="red", zlabel="eje z", label=false)
plot!(x1s,y1s,z1s, color="black", label=false)
title!("Ecuación paramétrica No. 2")
ylabel!("Eje Y")
xlabel!("Eje x")
```
